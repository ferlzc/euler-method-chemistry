```python
import numpy as np
from matplotlib import pyplot as plt

t0 = 0
tf = 50
n = 1000
deltax = (tf - t0)/(n - 1)

y10 = 1
y20 = 1
y30 = 1
y40 = 1
y50 = 1
y60 = 1

t = np.linspace(t0,tf,n)

y1 = np.zeros(n)
y2 = np.zeros(n)
y3 = np.zeros(n)
y4 = np.zeros(n)
y5 = np.zeros(n)
y6 = np.zeros(n)

y1[0]= y10
y2[0]= y20
y3[0]= y30
y4[0]= y40
y5[0]= y50
y6[0]= y60

k1 = 1
k2 = 1
k3 = 1
k4 = 1
k5 = 1
k6 = 1

def f1(x):
    return (-k1*y3[x-1]*y1[x-1] -k4*y6[x-1]*y1[x-1])

def f2(x):
    return (k2*y6[x-1]*y1[x-1])

def f3(x):
    return (-k1*y3[x-1]*y1[x-1]+k2*y4[x-1]+k4*y6[x-1])

def f4(x):
    return (k1*y3[x-1]*y1[x-1]-k2*y4[x-1]-k3*y4[x-1]*y5[x-1])

def f5(x):
    return (-k3*y5[x-1]*y4[x-1]+k4*y6[x-1]*y1[x-1])

def f6(x):
    return (k3*y4[x-1]*y5[x-1]-k4*y6[x-1]*y1[x-1])

for i in range(1,n):
    y1[i]= y1[i -1] + deltax*f1(i)  
    y2[i]= y2[i -1] + deltax*f2(i)
    y3[i]= y3[i -1] + deltax*f3(i)
    y4[i]= y4[i -1] + deltax*f4(i)
    y5[i]= y5[i -1] + deltax*f5(i)
    y6[i]= y6[i -1] + deltax*f6(i)
    
#for i in range ( n ):
#    print(t[i],y1[i])
```

```python
plt.plot(t,y1,'r+',t,y2,'b+',t,y3,'g+',t,y4,'r--',t,y5,'b--',t,y6,'g--' )
plt.xlabel(" t " )
plt.ylabel(" Concentração " )
plt.title(" Todos" )
plt.show()
```
